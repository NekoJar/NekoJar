```python
def norm(df) :
  df = (df - df.min()) / (df.max() - df.min())
  return df
```

```python
def euclidean(x1,x2) :
  return np.sqrt(np.sum((x1-x2)**2))
```

```python
euclidean(x.iloc[0], x.iloc[1])
```

```python
def knn(x_train, y_train, x_test, k) :

  dist = []

  for row in range(x_train.shape[0]) :
    dist.append(euclidean(x_train.iloc[row], x_test))

  data = x_train.copy()
  data['Jarak'] = dist
  data['Species'] = y_train
  data.sort_values(by='Jarak').reset_index(drop=True)

  y_pred = data.iloc[:k].Species.mode()

  return y_pred[0]
```

```python
def acc(y_pred, y_true) :
  true = 0
  for i in range(len(y_pred)):
    if y_pred[i] == y_true[i]:
      true += 1
  return true/len(y_pred)
```

```python
def evaluate(fold, k):
  test, train = fold
  x_train, y_train = train.drop('Species', axis = 1), train.Species
  x_test, y_test = test.drop('Species', axis = 1), test.Species

  x_train = norm(x_train)
  x_test = norm(x_test)

  y_pred = []

  for row in range(x_test.shape[0]): #iterasi loop melalui setiap baris df
    y_pred.append(knn(x_train, y_train, x_test.iloc[row], k))

  return (acc(y_pred, y_test))
```

```python
k = 5
accs = []
folds = [fold1, fold2, fold3] #fold1,fold2,fold3 menjadi 1 variable

for i in range(len(folds)) : #untuk melakukan iterasi melalui indeks dari list folds. Setiap iterasi, dilakukan evaluasi menggunakan fungsi evaluate() pada setiap fold dalam list folds, dengan menggunakan parameter k.
  accs.append(evaluate(folds[i], k))

print(f"k : {k}, dengan rata-rata akurasi: {sum(accs) / 3}")
```
